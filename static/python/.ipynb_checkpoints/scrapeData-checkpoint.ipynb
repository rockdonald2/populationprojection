```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpt
import matplotlib as mpl
import seaborn as sb
import requests
```

```python
codes = ['BE', 'BG', 'CZ', 'DK', 'DE', 'EE', 'IE', 'EL', 'ES', 'FR', 'HR', 'IT', 'CY', 'LV', 'LT', 'LU', 'HU', 'MT', 'NL', 'AT', 'PL', 'PT', 'RO', 'SI', 'SK', 'FI', 'SE', 'IS', 'LI', 'NO', 'CH']
```

```python
# http://ec.europa.eu/eurostat/wdds/rest/data/v2.1/json/en/
```

```python
population = []
```

```python
for c in codes:
    obj = {'code': c}
    year = 1990
    
    for _ in range(29):
        resp = requests.get('http://ec.europa.eu/eurostat/wdds/rest/data/v2.1/json/en/demo_r_d2jan?unit=NR&sex=T&age=TOTAL&time={}&geo={}'.format(year, c))
        
        obj['country'] = resp.json()['dimension']['geo']['category']['label'][c]
        
        obj[str(year)] = resp.json()['value']['0']
        
        year = year + 1
        
    population.append(obj)
```

```python
i = 0

for c in codes:
    resp = requests.get('http://ec.europa.eu/eurostat/wdds/rest/data/v2.1/json/en/proj_19np?unit=PER&sex=T&age=TOTAL&projection=BSL&geo={}'.format(c))
    
    year = 2019
    
    for v in resp.json()['value']:
        population[i][str(year)] = resp.json()['value'][v]
        year = year + 1

    i = i + 1
```

```python
df = pd.DataFrame(population)
df.loc[4, 'country'] = 'Germany'
df
```

```python
df = df.set_index('code')
```

```python
df.to_json('../data/population-data.json', orient='index')
```
