```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_json('../data/population-data.json', orient='index')
```

```python
df
```

```python
rows = 16
cols = 2

fig, axes = plt.subplots(ncols=cols, nrows=rows, figsize=(16, 80))

indices = df.index.values
i = 0
FLAG = False

for r in range(0, rows):
    for c in range(0, cols):
        df.loc[:, df.columns != 'country'].loc[indices[i]].plot(kind='line', ax=axes[r, c], title=indices[i], fontsize=12)
        i = i + 1
        
        if i == len(indices):
            FLAG = True
            break
            
    if FLAG:
        break

plt.tight_layout(pad=4.0)
plt.savefig('graph-comparison.jpg', dpi=200)
```
